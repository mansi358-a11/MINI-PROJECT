```python
# prompt: now as this code gives correct text from audio file recorded from a person diagnosed with dysarthria , now i want to convert it back to audio so that now it is clear

import pandas as pd
from google.colab import drive
import os
import librosa
import numpy as np
from transformers import pipeline
from gtts import gTTS
import IPython.display as ipd

# Mount Google Drive
drive.mount('/content/drive')

# Replace with your audio files folder path
audio_folder_path = '/content/drive/MyDrive/AudioFiles'

if os.path.exists(audio_folder_path):
  audio_files = []
  for filename in os.listdir(audio_folder_path):
    if filename.endswith(".mp3") or filename.endswith(".wav"):
      audio_files.append(os.path.join(audio_folder_path, filename))

  # Initialize the speech-to-text pipeline
  speech_to_text_pipeline = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h")

  # Initialize the BERT pipeline for sentiment analysis
  sentiment_pipeline = pipeline("sentiment-analysis")

  # Process each audio file
  for audio_file in audio_files:
    try:
      audio, sr = librosa.load(audio_file, sr=16000)
      text = speech_to_text_pipeline(audio)["text"]
      sentiment_result = sentiment_pipeline(text)

      print(f"Audio File: {audio_file}")
      print(f"Transcribed Text: {text}")
      print(f"Sentiment: {sentiment_result}")

      # Convert the transcribed text back to audio using gTTS
      tts = gTTS(text=text, lang='en')  # Adjust language if needed
      tts.save("output_audio.mp3")

      # Play the generated audio
      ipd.Audio("output_audio.mp3")

      print("-" * 30)

    except Exception as e:
      print(f"Error processing {audio_file}: {e}")

else:
  print("Audio folder not found at the specified path.")
```

```python
from google.colab import drive
import os
import librosa
from transformers import pipeline, BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch
from datasets import load_dataset

# **Mount Google Drive**
drive.mount('/content/drive')

# **Set path to shared audio folder in your Google Drive**
audio_folder_path = '/content/drive/MyDrive/speechocean/train'  # Replace with your actual path from the shared folder

# **Check if the folder exists and collect audio files**
if os.path.exists(audio_folder_path):
    audio_files = []
    for filename in os.listdir(audio_folder_path):
        if filename.endswith(".mp3") or filename.endswith(".wav"):  # Check for audio file types
            audio_files.append(os.path.join(audio_folder_path, filename))

    # **Load speech-to-text model**
    speech_to_text_pipeline = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h")

    # **Load BERT tokenizer and model for contextual and sentiment analysis**
    sentiment_pipeline = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

    # **Process each audio file**
    for audio_file in audio_files:
        try:
            # **Load the audio file using librosa**
            audio, sr = librosa.load(audio_file, sr=16000)  # Adjust sample rate if necessary

            # **Convert audio to text**
            text = speech_to_text_pipeline(audio)["text"]

            # **Perform sentiment and contextual analysis**
            sentiment_result = sentiment_pipeline(text)

            # **Output results**
            print(f"Audio File: {audio_file}")
            print(f"Transcribed Text: {text}")
            print(f"Sentiment: {sentiment_result}")
            print("-" * 30)

        except Exception as e:
            print(f"Error processing {audio_file}: {e}")

else:
    print("Audio folder not found at the specified path.")
```
